# Compute intersections on MultiLinestring

From ShapelyAddons Class :

```python
import os
import sys
sys.path.append(os.path.dirname(os.getcwd()))

from bokeh.plotting import output_notebook
from bokeh.plotting import show
from bokeh.io import curdoc, show

from shapely.wkt import loads as wkt_loads
from main_funcs import ExamplesCore
from main_funcs import BokehCore

output_notebook()

multilinestring = wkt_loads(
    'MULTILINESTRING Z ('
    '(-0.07199999999999999 0.85 0, -0.08 0.22 0),'
    '(-0.63 0.23 0, -0.08 0.22 0),'
    '(-0.08 0.22 0, 0.49 0.23 0),'
    '(0.65 0.47 0, 0.49 0.23 0),'
    '(-0.29 -0.07000000000000001 0, -0.08 0.22 0),'
    '(-0.37 -0.31 0, -0.29 -0.07000000000000001 0),'
    '(0.23 -0.13 0, -0.29 -0.07000000000000001 0)'
    ')'
)

geotools = ExamplesCore()
dead_ends = geotools.run_get_intersection_nodes(multilinestring, 'dead_ends')
crossroads = geotools.run_get_intersection_nodes(multilinestring, 'crossroads')
sections = geotools.run_get_intersection_nodes(multilinestring, 'sections')
```

```python
figure = BokehCore()

figure.add_multi_lines(multilinestring, 'MultiLineString', 'black')
figure.add_points(dead_ends, 'dead_ends', 'blue', size=8)
figure.add_points(crossroads, 'crossroads', 'red', size=8)
figure.add_points(sections, 'sections', 'green', size=8)

show(figure._plot)
```
